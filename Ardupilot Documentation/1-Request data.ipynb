```python
from pymavlink import mavutil
import subprocess
import time
import os
```

Kill related process

```python
for process in ["QGroundControl", "sim_vehicle.py", "arducopter", "gazebo", "mavproxy"]:
    os.system(f"pkill -9 -f {process}")
```

Launch Copter (ardupilot)

```python
sim_vehicle_path = os.path.expanduser("~/ardupilot/Tools/autotest/sim_vehicle.py")
vehicle_cmd = f"python3 {sim_vehicle_path} -v ArduCopter" 
# Open a new terminal and run the command
subprocess.Popen(["gnome-terminal", "--", "bash", "-c", f"{vehicle_cmd}; exec bash"])
```

Conect to the device

```python
conn = mavutil.mavlink_connection('udp:127.0.0.1:14550')

conn.wait_heartbeat()
print("Heartbeat received. Connected to the vehicle.")
```

## Request a Message

[MAV_CMD_REQUEST_MESSAGE ](https://mavlink.io/en/messages/common.html#MAV_CMD_REQUEST_MESSAGE)

Examples
This messages is automatically broadcasted anyway

```python
conn.mav.command_long_send(
    conn.target_system, conn.target_component,
    mavutil.mavlink.MAV_CMD_REQUEST_MESSAGE, 0,
    mavutil.mavlink.MAVLINK_MSG_ID_GLOBAL_POSITION_INT, 0, 0, 0, 0, 0, 0
)
```

## Listen to messages

These are pymavlink comands.

Listen to any mesaage

```python
start_time = time.time()  # Get the current time
print("hearing...")
while time.time() - start_time < 1:  # Run for 0.01 seconds
    msg=conn.recv_msg()
    if msg is not None:
        print(msg)
```

Listen to a specific message(blocking execution)

```python
msg=conn.recv_match(type='ATTITUDE',blocking=True)

print(msg)
```

```python
msg=conn.recv_match(type='EXTENDED_SYS_STATE', blocking=False) 
# Try this for get stucked 
#conn.recv_match(type='EXTENDED_SYS_STATE', blocking=True) 
print(msg)
```
